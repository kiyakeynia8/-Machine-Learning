```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
apple_width = np.random.normal(6, 1, 100)
apple_length = np.random.normal(6, 1, 100)
banana_width = np.random.normal(4, 1, 100)
banana_length = np.random.normal(8, 1, 100)
watermelon_width = np.random.normal(10, 1, 100)
watermelon_length = np.random.normal(10, 1, 100)
```

```python
x_width = 3
x_length = 9

y_width = 7
y_length = 5

z_width = 8
z_length = 8
```

```python
plt.scatter(apple_width, apple_length, c= "red", marker= ".")
plt.scatter(banana_width, banana_length, c= "orange", marker= ".")
plt.scatter(watermelon_width, watermelon_length, c= "green", marker= ".")
plt.scatter(x_width, x_length, c= "blue", marker= "o")
plt.scatter(y_width, y_length, c= "green", marker= "o")
plt.scatter(z_width, z_length, c= "black", marker= "o")

plt.show()
```

```python
width = np.concatenate((apple_width, banana_width, watermelon_width))
length = np.concatenate((apple_length, banana_length, watermelon_length))

X_train = np.array([width, length]).T
# X_train
```

```python
Y_apples = np.zeros(100, dtype="int")
Y_bananas = np.ones(100, dtype="int")
Y_watermelon = np.ones(100, dtype="int")*2
Y_train = np.concatenate((Y_apples, Y_bananas, Y_watermelon))
# Y_train
```

```python
def euclidean(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))
```

```python
new_fruit = np.array([z_width, z_length])

distances = []
for x in X_train:
    d = euclidean(new_fruit, x)
    distances.append(d)
```

```python
nearrest_neighbors = np.argsort(distances)[0:5]
nearrest_neighbors
```

```python
result = np.bincount(Y_train[nearrest_neighbors])
```

```python
a_b = np.argmax(result)
```

```python
plt.scatter(apple_width, apple_length, c= "red", marker= ".")
plt.scatter(banana_width, banana_length, c= "orange", marker= ".")
plt.scatter(watermelon_width, watermelon_length, c= "green", marker= ".")
plt.scatter(x_width, x_length, c= "blue", marker= "o")
plt.scatter(y_width, y_length, c= "green", marker= "o")

if a_b == 0:
    plt.scatter(z_width, z_length, c= "red", marker= "p")
elif a_b == 1:
    plt.scatter(z_width, z_length, c= "orange", marker= "p")
elif a_b == 2:
    plt.scatter(z_width, z_length, c= "green", marker= "p")

plt.show()
```

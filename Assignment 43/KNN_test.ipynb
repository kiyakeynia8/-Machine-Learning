```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from KNN_main import KNN
```

```python
def generate_dataset(n):
    apple_width = np.random.normal(6, 1, n)
    apple_length = np.random.normal(6, 1, n)
    banana_width = np.random.normal(4, 1, n)
    banana_length = np.random.normal(8, 1, n)
    watermelon_width = np.random.normal(10, 1, n)
    watermelon_length = np.random.normal(10, 1, n)
    width = np.concatenate((apple_width, banana_width, watermelon_width))
    length = np.concatenate((apple_length, banana_length, watermelon_length))
    X = np.array([width, length]).T
    Y_apples = np.zeros(n, dtype="int")
    Y_bananas = np.ones(n, dtype="int")
    Y_watermelon = np.ones(n, dtype="int")*2
    Y = np.concatenate((Y_apples, Y_bananas, Y_watermelon))

    return X, Y
```

```python
x_train, y_train = generate_dataset(100)
```

```python
x_width = 3
x_length = 9

y_width = 7
y_length = 5
```

```python

cmap = ["red", "orange", "green"]
plt.scatter(x_train[:, 0], x_train[:, 1],c= y_train, cmap=ListedColormap(cmap), marker= ".")

plt.show()
```

```python
knn = KNN(5)
knn.fit(x_train, y_train)
```

```python
new_fruit = np.array([3, 5])
new_fruit_2 = np.array([6, 5.5])

new_fruits = [new_fruit, new_fruit_2]

output = knn.predict(new_fruits)
```

```python
cmap = ["red", "orange", "green"]
plt.scatter(x_train[:, 0], x_train[:, 1],c= y_train, cmap=ListedColormap(cmap), marker= ".")

for y in output:
    if y == 0:
        print("apple")
    elif y == 1:
        print("banana")
    elif y == 2:
        print("watermelon")
plt.show()
```

```python
X_test, Y_test = generate_dataset(90)
```

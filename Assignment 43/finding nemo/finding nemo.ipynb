```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

from KNN_main import KNN
```

```python
nemo = cv2.imread("nemo.jpg")
nemo = cv2.cvtColor(nemo, cv2.COLOR_BGR2RGB)
nemo = cv2.resize(nemo, (0, 0), fx= 0.25, fy=0.25)
plt.imshow(nemo)
plt.show()
```

```python
pixels_list_rgb = nemo.reshape(-1, 3)

pixels_list_rgb.shape

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection = "3d")

ax.scatter(pixels_list_rgb[:, 0], pixels_list_rgb[:, 1], pixels_list_rgb[:, 2], c = pixels_list_rgb/255)
plt.show()
```

```python
nemo_hsv = cv2.cvtColor(nemo, cv2.COLOR_RGB2HSV)
```

```python
pixels_list_hsv = nemo_hsv.reshape(-1, 3)

pixels_list_hsv.shape

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection = "3d")

ax.scatter(pixels_list_hsv[:, 0], pixels_list_hsv[:, 1], pixels_list_hsv[:, 2], c=pixels_list_rgb/255)
plt.show()
```

```python
light_orange = (1, 100, 100)
dark_orange = (60, 255, 255)

mask = cv2.inRange(nemo_hsv, light_orange, dark_orange)
plt.imshow(mask, cmap="gray")
plt.show()
```

```python
light_white = (0, 0, 150)
dark_white = (145, 60, 255)

mask_white = cv2.inRange(nemo_hsv, light_white, dark_white)
plt.imshow(mask_white, cmap="gray")
plt.show()
```

```python
light_black = (0, 0, 0)
dark_black = (255, 255, 50)

mask_black = cv2.inRange(nemo_hsv, light_black, dark_black)
plt.imshow(mask_black, cmap="gray")
plt.show()
```

```python
final_mask = mask + mask_white + mask_black
plt.imshow(final_mask, cmap="gray")
plt.show()
```

```python
final_result = cv2.bitwise_and(nemo, nemo, mask = final_mask)
plt.imshow(final_result)
plt.show()
```

```python
X_train = pixels_list_hsv / 255
X_train
```

```python
Y_train = final_mask.reshape(-1,) // 255
Y_train
```

```python
knn = KNN(3)
knn.fit(X_train, Y_train)
```

```python
new_pixel = np.array([255, 80, 0])/255
knn.predict([new_pixel])
```

```python
dadash_nemo = cv2.imread("dadash_nemo.jpg")
dadash_nemo = cv2.cvtColor(dadash_nemo, cv2.COLOR_BGR2RGB)
dadash_nemo = cv2.resize(dadash_nemo, (0, 0), fx= 0.2, fy=0.2)
plt.imshow(dadash_nemo)
plt.show()
```

```python
pixels_list_hsv = dadash_nemo.reshape(-1, 3)
X_test = pixels_list_hsv / 255
Y_pred = knn.predict(X_test)
Y_pred
```

```python
output = np.array(Y_pred).reshape(dadash_nemo.shape[:2])
plt.imshow(output, cmap="gray")
plt.show()
```

```python
output = output.astype("uint8")
```

```python
final_result = cv2.bitwise_and(dadash_nemo, dadash_nemo, mask=output)
plt.imshow(final_result)
plt.show()
```

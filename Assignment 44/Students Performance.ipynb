```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from scipy import stats
from LLS_main import LLS
```

```python
data_X = np.random.uniform(0, 8 , 1000)
data_Y = 2*data_X + np.random.normal(0,2,1000) + 2
```

```python
plt.scatter(data_X, data_Y, color= 'green', marker='.')
plt.xlabel("Hours")
plt.ylabel("Grade")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data_X, data_Y, test_size = 0.2)
# X_train = np.array(X_train)
# X_test = np.array(X_test)
# Y_train = np.array(Y_train)
# Y_test = np.array(Y_test)
```

```python
X_train = X_train.reshape(-1, 1)
Y_train = Y_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
Y_test = Y_test.reshape(-1, 1)
```

```python
lls = LLS()
lls.fit(X_train, Y_train)
Y_pred = lls.predict(X_test)
```

```python
slope, _, _, _, _ = stats.linregress(data_X, data_Y)
```

```python
Y_pred
```

```python
plt.scatter(X_test, Y_test, marker=".", color="red")
plt.plot(X_test, Y_pred, color="blue")
plt.plot(X_test, Y_pred * 2, color="red")
plt.xlabel("Study Hours")
plt.ylabel("Grade")
plt.show()

plt.scatter(X_train, Y_train, marker=".", color="blue")
plt.plot(X_test, slope * X_test, color="green")
plt.xlabel("Study Hours")
plt.ylabel("Grade")
plt.show()
```
